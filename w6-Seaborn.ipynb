### Bar Plot (Çubuk Grafiği)

Kategorik değişkenleri görselleştirmek için kullanmaktayız.

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
tips.describe()
```

```python
tips.dtypes
```

```python
tips.groupby('day')['total_bill'].agg(['size','min', 'max', 'mean', 'median', 'var'])
```

```python
sns.set(style="whitegrid")

plt.figure( figsize=(10, 5) )

ax = sns.barplot(x="day", y="total_bill", data=tips)

plt.ylabel('Total Bill')
plt.xlabel('Days')
plt.title('Total Bill Average by Day')

plt.show()
```

```python
plt.figure( figsize=(12, 6) )

ax = sns.barplot(x="day", y="total_bill", hue="sex", data=tips)

plt.ylabel('Total Bill')
plt.xlabel('Days')

plt.title('Total Bill Average by Day and Sex')

plt.show()
```

```python
plt.figure( figsize=(12, 6) )

ax = sns.barplot(x="time", y="tip", data=tips, 
                 order=["Dinner", "Lunch"])
plt.show()
```

```python
plt.figure( figsize=(12, 6) )

ax = sns.barplot(x="city", y="salary", data=df_data, ci=95)

plt.ylabel('Salary')
plt.xlabel('City')

plt.title('Salary Average by City and Gender')

plt.show()
```

```python
plt.figure( figsize=(12, 6) )

ax = sns.barplot(x="city", y="salary", hue="gender", data=df_data, ci=95)

plt.ylabel('Salary')
plt.xlabel('City')

plt.title('Salary Average by City and Gender')

plt.show()
```

### Line Plot (Çizgi Grafiği)


```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.describe()
```

```python
fmri.dtypes
```

```python
ax = sns.lineplot(x="timepoint", y="signal", data=fmri, ci=0)
```

```python
dates = pd.date_range(start="01/01/2019", end='31/12/2019', freq='D')
udata = np.random.uniform(100, 500, 365)

df_ts = pd.DataFrame({'Date': dates, 'Uniform': udata })

df_ts
```

```python
plt.figure( figsize=(20, 5) )
ax = sns.lineplot(x="Date", y="Uniform", 
                  data=df_ts)

plt.xlabel('Monhts')
plt.ylabel('Value of Uniform Distribution')
plt.title('Time Series')
plt.xticks(pd.date_range(start="01/01/2019", end='31/12/2019', freq='M'), rotation=90)
plt.yticks()


plt.show()
```

### Scatter Plot

Dağılım Grafiği iki sürekli değişken arasındaki ilişkiyi görmek için kullanılır.

Scatter Plot is used to see the relationship between two continuous variables.

```python
tips = sns.load_dataset("tips")


ax = sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
ax = sns.scatterplot(x="total_bill", y="tip", hue="time",
                     data=tips)
```

### Box Plot (Box and Whisker Plot)

```python
tips = sns.load_dataset("tips")

ax = sns.boxplot(x=tips["total_bill"], orient='h')
```

```python
ax = sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
ax = sns.boxplot(x="day", y="total_bill", hue="smoker",
                 data=tips)
```

```python
ax = sns.boxplot(x="time", y="tip", data=tips,
                 order=["Dinner", "Lunch"])
```

```python
iris = sns.load_dataset("iris")
ax = sns.boxplot(data=iris, orient="h", palette="Set3")
```

### HeatMap

```python
flights = sns.load_dataset("flights")

flights = flights.pivot("month", "year", "passengers")

flights
```

```python
ax = sns.heatmap(flights)
```

```python
plt.figure( figsize=(12, 6) )

ax = sns.heatmap(flights, annot=True, fmt="d")

plt.show()
```
